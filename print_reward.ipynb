```python
import json
import matplotlib.pyplot as plt

# 加载 jsonl 文件
jsonl_file = "/home/zhangzhibo/logdir/20241117T135935-example/metrics.jsonl"
rewards = []

with open(jsonl_file, 'r') as f:
    for line in f:
        data = json.loads(line)
        if "episode/score" in data:
            rewards.append(data["episode/score"])

# 可视化奖励曲线
plt.plot(rewards)
plt.xlabel("Episodes")
plt.ylabel("Cumulative Reward")
plt.title("Training Reward Curve")
plt.show()
```

```python
import imageio
import numpy as np

# 假设 npz_file 是你已经加载的 .npz 文件
npz_file = np.load('/home/zhangzhibo/logdir/20241117T135935-example/replay/20241119T203228F757592-6TuaklpWyuLrrD3F8UcPT0-6rL5N8Lmw6Pwpko9sSA3gc-192.npz')
frames = npz_file['walker_egocentric_camera']

# 确定输出视频的帧率
fps = 30  # 你可以根据需要调整帧率

# 使用 imageio.get_writer 写入视频
output_path = 'video.mp4'
with imageio.get_writer(output_path, fps=fps) as writer:
    for frame in frames:
        writer.append_data(frame)

print(f"Video saved to {output_path}")
```
